```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd

iris = datasets.load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

cell = iris.feature_names

df_train = pd.DataFrame(X_train, columns=cell)
df_train['target'] = y_train
df_train['dataset'] = 'train'

df_test = pd.DataFrame(X_test, columns=cell)
df_test['target'] = y_test
df_test['dataset'] = 'test'

df_combined = pd.concat([df_train, df_test], ignore_index=True)

print(df_combined)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(9, 7))
sns.scatterplot(data=df_combined, x='petal length (cm)', y='petal width (cm)', hue='target', style='dataset', palette='muted')

plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

feature_x = 'petal length (cm)'
feature_y = 'petal width (cm)'
feature_z = 'sepal length (cm)'

colors = {0: 'purple', 1: 'orange', 2: 'green'}

fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(projection='3d')

for target, color in colors.items():
    subset = df_combined[df_combined['target'] == target]
    ax.scatter(subset[feature_x], subset[feature_y], subset[feature_z],
               c=color, label=iris.target_names[target], s=25, alpha=0.5)

ax.set_xlabel(feature_x)
ax.set_ylabel(feature_y)
ax.set_zlabel(feature_z)
ax.legend(title='iris type')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


for column in df_combined.columns:
    plt.figure(figsize=(10, 6))
    sns.heatmap(df_combined[[column]], cmap='coolwarm', annot=False, cbar=True)
    plt.title(f'Heatmap for {column}')
    plt.show()
```

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

plt.figure(figsize=(12, 10))

for i in range(len(features)):
    feature = features[i]

    plt.subplot(2, 2, i + 1)

    sns.kdeplot(df_combined[df_combined['dataset'] == 'train'][feature],
                label='Train', color='blue', fill=True, alpha=0.4)

    sns.kdeplot(df_combined[df_combined['dataset'] == 'test'][feature],
                label='Test', color='red', fill=True, alpha=0.4)

    plt.title(f'Density Plot of {feature}')
    plt.xlabel(feature)
    plt.legend()

plt.suptitle('Density Plots of Iris Dataset Features (Train vs Test)', fontsize=14, y=1.02)
plt.show()
```
