# Defining the model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define activation functions
def hardlim(x):
    return 1 if x >= 0 else 0
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Defining McCulloch-Pitts neuron class
class McCulloch_Pitts_Neuron:
    def __init__(self, weights, bias, activation_func):
        self.weights = np.array(weights)
        self.bias = bias
        self.activation_func = activation_func

    def model(self, x):
        # Compute the net input
        n = np.dot(self.weights, x) + self.bias
        # Apply activation function
        return self.activation_func(n)
```

# Using hardlim activation function

```python
# Define the Area function
def Area_hardlim(x, y):
    # Neurons for edges with sigmoid activation
    neur1 = McCulloch_Pitts_Neuron([-2, -1], 6, hardlim)  # AB
    neur2 = McCulloch_Pitts_Neuron([0, 1], 0, hardlim)      # BC
    neur3 = McCulloch_Pitts_Neuron([2, -1], -2, hardlim)     # CA
    neur4 = McCulloch_Pitts_Neuron([1, 1, 1], -3, hardlim)

    # Calculate neuron outputs
    z1 = neur1.model([x, y])
    z2 = neur2.model([x, y])
    z3 = neur3.model([x, y])
    z4 = neur4.model([z1, z2, z3])

    return list([z4])
```

```python
# Generate random data points
num_points = 2000
x_values = np.random.uniform(0, 4, num_points)  # Updated x-axis limits
y_values = np.random.uniform(-1, 3, num_points)  # Updated y-axis limit
```

```python
# Initialize lists to store data points for different z4 values
red_points = []
green_points = []

# Evaluate data points using the Area function
for i in range(num_points):
    z4_value = Area_hardlim(x_values[i], y_values[i])
    if z4_value == [0]:  # z4 value is 0
        red_points.append((x_values[i], y_values[i]))
    else:  # z4 value is 1
        green_points.append((x_values[i], y_values[i]))

# Separate x and y values for red and green points
red_x, red_y = zip(*red_points)
green_x, green_y = zip(*green_points)
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(red_x, red_y, color='red', label='z5 = 0')
plt.scatter(green_x, green_y, color='green', label='z5 = 1')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('McCulloch-Pitts Neuron Outputs')

# Plot triangle edges
A = [2, 2]
B = [3, 0]
C = [1, 0]

plt.plot([A[0], B[0]], [A[1], B[1]], 'k--')
plt.plot([B[0], C[0]], [B[1], C[1]], 'k--')
plt.plot([C[0], A[0]], [C[1], A[1]], 'k--')
plt.grid(True)

# Set axis limits
plt.xlim(0, 4)
plt.ylim(-1, 3)

# Position the legends at the top and right
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

# Save plot as PDF
plt.savefig('c.png', bbox_inches='tight')

plt.show()
```
